```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set=pd.read_csv("Salary_dataset.csv")
data_set
```

**x = YearsExperience.................(FEATURE USED TO PROTECT THE TARGET)**

**y = Salary...............(TARGET)**


```python
x = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
#Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Fitting the impler linear Regression model to the traning dataset
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Prediction of test and training set result
y_pred = regressor.predict(x_test)
x_pred = regressor.predict(x_train)
```

```python
y_pred
```

```python
x_pred
```

```python
#checking  the accuracy :
train_score = regressor.score(x_train,y_train)
test_score = regressor.score(x_test,y_test)

print("Score of the train set is : ",train_score)
print("Score of the test set is : ",test_score )
```

```python
#OUTPUT PREDICTION


#entering inputs :
Years  =float(input("Enter the  Year of Experience : "))

#building a predictive system:
arr= [Years]
input1= arr

#changing the input data into numpy array:
input1_as_numpy_array = np.asarray(input1)

#Reshaping the input :
input1_reshaped = input1_as_numpy_array.reshape(1,-1)

#Predicting the output :
prediction = regressor.predict(input1_reshaped)

#displaying the prediction
Salary = float(prediction)
print("THE SALARY IS : ", Salary ,"rupees")


```

```python
#Visualizing the training set results

plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,x_pred, color = "red")
plt.title("Salary Vs Experience (Training Set)")
plt.xlabel("Yeas of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_train, y_train, color="green", label="Training Data")  # Green for training data
plt.plot(x_train, x_pred, color="blue", label="Prediction Line")    # Blue for prediction line
plt.title("Salary Vs Experience (Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()  # Add a legend to distinguish colors
plt.show()
```
